```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

uncleaned_data = pd.merge(study_results, mouse_metadata)

uncleaned_data.head()
```

```python
uncleaned_num_of_mice = uncleaned_data["Mouse ID"].nunique()
uncleaned_num_of_mice
```

```python
duplicates = uncleaned_data[uncleaned_data.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
duplicates_by_mouse_id = duplicates["Mouse ID"].unique()
duplicates_by_mouse_id
```

```python
uncleaned_data[uncleaned_data["Mouse ID"].isin(duplicates_by_mouse_id)]
```

```python
data = uncleaned_data[~uncleaned_data["Mouse ID"].isin(duplicates_by_mouse_id)]
data.head()
```

```python
num_of_mice = data["Mouse ID"].nunique()
num_of_mice
```
